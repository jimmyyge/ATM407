### Background
**Kinematics** is not dynamics, it is just a *description* of how 2D vector fields vary in 2D. 

* Assume LINEAR dependence of u,v on x,y 
    * This involves 4 coefficients $c_1, c_2, c_3, c_4$
* Also, add a constant (not variying with x,y) vector field U,V 

$$ u = U + c_1 x  +c_2 y$$
$$ v = V + c_3 x  +c_4 y$$

## Exercise 1 Answer:

* $\newcommand{\pd}[2]{\frac{\partial{#1}}{\partial{#2}}}
div = \pd{u}{x} + \pd{v}{y}= \pd{}{x}(U + c_1x + c_2y) + \pd{}{y}(V + c_3x + c_4y) = \pd{}{x}(c_1x) + \pd{}{y}(c_4y) = c_1 + c_4$
* $ \zeta_{rel} = \pd{v}{x} - \pd{u}{y} = \pd{}{x}(V + c_3x + c_4y) - \pd{}{y}(U + c_1x + c_2y) = \pd{}{x}(c_3x) - \pd{}{y}(c_2y) = c_3 - c_2 $
* $ def1 = \pd{u}{x} - \pd{v}{y} = \pd{}{x}(c_1x) - \pd{}{y}(c_4y) = c_1 - c_4 $
* $ def2 = \pd{v}{x} + \pd{u}{y} = \pd{}{x}(c_3x) + \pd{}{y}(c_2y) = c_3 + c_2 $

## Exercise 2: Create and display the basic fields
* First, execute a code cell setting up a coordinate system (x,y) from -10 to 10
* Next, execute the cell defining the basic "pure" div, vort, def1, def2, and U,V fields 
* Finally, display them. The first one has been done for you. You complete the work. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Set up a coordinate grid centered on 0,0
n = 10
x,y = np.mgrid[-n:n+1, -n:n+1]
```

```python
# Pure divergence of unit magnitude
udiv,vdiv = 1*x, 1*y

# Pure vorticity of unit magnitude
uvor,vvor = -1*y, 1*x

# Pure def1 of unit magnitude
udef1,vdef1 = 1*x, -1*y

# Pure def2 of unit magnitude
udef2,vdef2 = 1*y, 1*x

# Pure U and V
uU, vU = 1, 0
uV, vV = 0, 1
```

```python
# Define the display field udisp,vdisp and give it a label. 
# Edit-> Copy and paste this cell, then adjust the top 2 lines for the other cases

udisp,vdisp = 2*uU + udiv, 2*vU + vdiv
label = 'Pure unit divergence plus U=2'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()


udisp,vdisp = 2*uU + uvor, 2*vU + vvor
label = 'Pure unit vorticity plus U=2'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()


udisp,vdisp = 2*uU + udef1, 2*vU + vdef1
label = 'Pure unit deformation 1 plus U=2'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()


udisp,vdisp = 2*uU + udef2, 2*vU + vdef2
label = 'Pure unit deformation 2 plus U=2'

# Display the vector field as arrows, CENTERED ON the point they apply to
fig, ax = plt.subplots(figsize=(7,7))
plt.quiver(x, y, udisp, vdisp, pivot='mid')
plt.title(label)
plt.show()
```

## Answer to Exercise 3: 

Vorticity plus deformation (two curved flow patterns) makes a flow we could describe as shearing, with two adjacent flow regions moving antiparallel to each other. In deformation 1, there is a skew to the flow field (the shearing "axis" is diagonal), whereas in deformation 2, the shearing is vertical. This is interesting because in the deformation fields _without_ vorticity, the deformation 1 flow was vertical and horizontal and never crossed the _x_ or _y_ axes, while deformation 2 had a diagonal element.

## Answer to Exercise 4: 

Vorticity plus straight flow can be described as cyclonic with the center of the flow displaced upwards as _U_ increases. (The increased _U_ counteracts the counterclockwise vorticity flow above the center.) At lower levels, the cyclone rotates about the center of the grid. At upper levels, the stronger upper-level westerlies displace the center of rotation to the north, resulting in a tilted vertical profile of the system.

## Answer to Exercise 5: 

When you combine vorticity and convergence (negative divergence), you get a pattern of cyclonic flow with the vectors pointed slightly inward to the center of the field, as opposed to perfectly tangent around the flow. This is similar to what is seen at the surface in low pressure systems, where friction slows the wind and disrupts geostrophic balance, causing air to spiral inward to the center of the low. One thing to point out here is that the magnitude of the vectors becomes smaller nearer the center, whereas in the real world, wind speeds increase closer to the center due to the increasing pressure gradient.
