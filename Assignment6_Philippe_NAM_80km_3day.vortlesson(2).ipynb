### Background
Recall the spirit of the *vorticity equation*: It eliminated the PGF from the horizontal momentum equations:  

* Horizontal acceleration equation (F=ma):
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu $$

* Eliminate $\Phi$ by subtracting the y derivative of the u equation from the x derivative of the v equation: 

$$ \frac{d}{dt}(\zeta_{abs}) = 0 - \zeta_{abs} (\nabla \cdot V)$$

* Recall that $ \zeta_{abs} = \zeta_{rel} + f $, where f is *planetary vorticity*.

* Divergence changes the ***horizontal area A*** of a mass of a blob of air the same way. 

$$ \frac{1}{A} \frac{dA}{dt} = (\nabla \cdot V) $$

$$ \frac{1}{\zeta_{abs}} \frac{d \zeta_{abs}}{dt} = (\nabla \cdot V) $$

* If the air blob's mass is constant, but A increases, then its depth $\delta p$ decreases. 

$$ \frac{1}{\delta p} \frac{d\delta p}{dt} = -(\nabla \cdot V) $$

* But what is this "blob" of air? It is bounded by two material surfaces -- **such as $\theta$ surfaces**. So the equation above refers to *the $\delta p$ for a given $\delta \theta$*. 

* We can therefore eliminate the divergence term: The ratio stays constant, since convergence *amplifies the numerator* but *decays the denominator* by the same fractional amount. 

$$ \frac{d}{dt}(PV) = 0$$

$$ PV = \frac {\zeta_{abs}} {(\delta p / \delta \theta)}$$

(Often people toss in the constant factor of g to turn $\delta p$ (a *weight* of air) into a *mass*.) 


* Except that $\theta$ isn't quiiite conserved... so we need to write 

$$ \frac{d}{dt}(PV) = \zeta_{abs} \frac {\partial}{\partial p}(d\theta/dt) = \zeta_{abs} \frac {\partial}{\partial p}(Q)$$

where Q is our catch-all letter for "diabatic" heating rate (here expressed in $\theta$ units, although that is still K/s or K/d units).
