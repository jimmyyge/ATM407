### Background and math

You will decompose the total wind into geostrophic and ageostrophic aspects: $\vec V = \vec V_g + \vec V_a$. 

* Horizontal acceleration equations (F=ma) for inviscid flow:
$$ \frac{du}{dt} = - \frac{\partial \Phi}{\partial x} + fv $$

$$ \frac{dv}{dt} = - \frac{\partial \Phi}{\partial y} - fu $$

* The geostrophic wind is defined as *the wind that would occur if the LHS were zero*: 

$$ 0 = - \frac{\partial \Phi}{\partial x} + fv_g $$

$$ 0 = - \frac{\partial \Phi}{\partial y} - fu_g $$

## Exercise 1 Answer:

$$
\frac{\partial}{\partial y} \left[ 0 = -\frac{\partial \Phi}{\partial y} + fv_g \right] - \frac{\partial}{\partial x} \left[ 0 = -\frac{\partial \Phi}{\partial y} - fu_g \right]$$

$$0 + 0 = -\frac{\partial^2 \Phi}{\partial x \partial y} + v_g \frac{\partial f}{\partial y} + f \frac{\partial v_g}{\partial y} + \frac{\partial^2 \Phi}{\partial y \partial x} + u_g \frac{\partial f}{\partial x} + f \frac{\partial u_g}{\partial x}$$

$$0 = u_g \frac{\partial f}{\partial x} + v_g \frac{\partial f}{\partial y} + f \frac{\partial u_g}{\partial x} + f \frac{\partial v_g}{\partial y}$$

$$\frac{\partial f}{\partial x} = 0; \frac{\partial f}{\partial y} = \beta$$

$$0 = \frac {v_g \beta}{f} + \left( \frac{\partial u_g}{\partial x} + \frac{\partial v_g}{\partial y} \right)$$

If $\beta = 0$:

$$\left( \frac{\partial u_g}{\partial x} + \frac{\partial v_g}{\partial y} \right) = 0$$

#### Key points: the geostrophic wind is boring for 2 reasons: because it is not connected to vertical velocity $\omega$, and also because it would never change with time since its definition sets d/dt to zero. The implication is that ageostrophic wind, despite its relatively small magnitude, is especially *interesting*. However, the geostrophic wind is adequate to approximate the direction and speed of advection of various scalar fields (temperature, vorticity) in the atmosphere.

## Answer to exercise 3: Acceleration-related ageostrophic winds 

I chose to focus on a cyclone off the coast of British Columbia, and a parcel located slightly south of the cyclone's center. The Coriolis force (blue arrow) acts to the right, i.e. northward, of the westward ageostrophic wind (green arrow), resulting in an northward acceleration of the parcel directed toward the center of the low.

Since geostrophic wind does not change in time, we have that the acceleration $\frac{dV}{dt} = f \vec{V_a} \times \hat{k}$. Since $\vec{V_a}$ is travelling in the negative $\hat{i}$ (west) direction, the cross product with $\hat{k}$ (upwards) results in a vector in the positive $\hat{j}$ direction (northwards). The magnitude of the acceleration depends only on the magnitudes of the Coriolis parameter, $f$, and $\vec{V_a}$.
