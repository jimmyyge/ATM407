### Background
The primitive equations:
1. Horizontal acceleration equation (F=ma):
$$ \frac{D}{Dt} \vec V_h = -f \hat k \times \vec V_h - \vec \nabla_p \Phi +\vec F_r  $$
or, in component form, 
$$ \frac{Du}{Dt} = fv  - \frac{\partial \Phi}{\partial x} +F $$
$$ \frac{Dv}{Dt} = -fu - \frac{\partial \Phi}{\partial y} +G $$

1. Vertical acceleration equation (aka Hydrostatic balance): 
$$ \frac {\partial \Phi}{\partial p} = - \alpha \equiv -\frac{1}{\rho} = \frac{RT}{p} $$
1. Mass continuity: 
$$ 0 = \frac {\partial u}{\partial x} + \frac {\partial v}{\partial y}  + \frac {\partial \omega}{\partial p} $$
1. First Law of thermodynamics: 
$$ \frac {D_h T}{Dt} + \omega \frac {\partial \theta}{\partial p} = J/C_p $$


Now suppose the flow is in geostrophic balance (dV/dt=0, F=G=0). This defines the geostrophic wind:  
$$ fv_g = +\frac{\partial \Phi}{\partial x}$$
$$ fu_g = -\frac{\partial \Phi}{\partial y}$$


Taking the p derivative, and noticing that $\frac{\partial \Phi}{\partial p} = g\frac{\partial Z}{\partial p}$ is the *geometric thickness* (dZ) of a layer with mass (dp/g), 
$$ \frac {\partial}{\partial p}v_g = +g/f \frac{\partial \frac {\partial Z}{\partial p}}{\partial x}$$
$$ \frac {\partial}{\partial p}u_g = -g/f \frac{\partial \frac {\partial Z}{\partial p}}{\partial y}$$

The horizontal pattern of thickness is the *pattern of temperature*, as you know. But actually plugging in the hydrostatic equation above for $\frac{\partial \Phi}{\partial p}$ is complicated, because p is in the denominator of the right hand side. So you will do an exercise below (on paper, then type in the answer) using log-pressure as a "pseudo-height" coordinate.

## Exercise 4 answer: 

The thermal wind is parallel to temperature contours, so it does *not* advect temperature. Since $V_{thermal} \cdot \nabla T = 0$, the temperature advection is the same at 500 hPa and 850 hPa.
\begin{align}
V_{g,lower} + V_{thermal} &= V_{g,upper}\\
(V_{g,850} + V_T) \cdot \nabla T &= V_{g,500} \cdot \nabla T \\
V_{g,850} \cdot \nabla T + V_T \cdot \nabla T &= V_{g,500} \cdot \nabla T\\
V_{g,850} \cdot \nabla T + 0 &= V_{g,500} \cdot \nabla T\\[10pt]
V_{g,850} \cdot \nabla T &= V_{g,500} \cdot \nabla T\\
\end{align}
